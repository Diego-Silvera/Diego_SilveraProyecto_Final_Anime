## 5. Procesamiento NLP (normalizaci√≥n)
**Objetivo:** normalizar texto (min√∫sculas, signos, stopwords, stemming/lemma).  
**Resultado esperado:** columna `clean_text` lista para vectorizaci√≥n (TF-IDF).

```python
import nltk, re
from nltk.corpus import stopwords
from nltk.stem.snowball import EnglishStemmer
# nltk.download('stopwords')

STOP_EN = set(stopwords.words('english'))
stemmer = EnglishStemmer()

def clean_basic(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", " ", text)
    text = re.sub(r"[^a-z\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

def normalize(text):
    toks = [w for w in clean_basic(text).split() if w not in STOP_EN and len(w) > 2]
    toks = [stemmer.stem(w) for w in toks]
    return " ".join(toks)

if 'synopsis' in df.columns:
    df['clean_text'] = df['synopsis'].apply(normalize)
else:
    df['clean_text'] = ""

df[['title','clean_text']].head(3) if 'title' in df.columns else df[['clean_text']].head(3)
```

#Conclusi√≥n ‚Äì Procesamiento NLP (normalizaci√≥n)

La etapa de normalizaci√≥n textual permiti√≥ transformar el corpus original de sinopsis en un formato uniforme, limpio y adecuado para su vectorizaci√≥n, asegurando la coherencia sem√°ntica de los textos y reduciendo la presencia de ruido ling√º√≠stico.
El proceso incluy√≥ la conversi√≥n a min√∫sculas, la eliminaci√≥n de signos de puntuaci√≥n y URLs, el filtrado de stopwords y la aplicaci√≥n de stemming mediante el algoritmo de Snowball.
Estas operaciones garantizaron que las palabras conservadas representaran √∫nicamente la informaci√≥n relevante del mensaje, eliminando elementos redundantes o irrelevantes para el an√°lisis.

El resultado de este proceso fue la creaci√≥n de la columna clean_text, la cual constituye la base del modelado posterior.
Su estructura refleja un vocabulario depurado y homog√©neo, con tokens estandarizados que permiten una representaci√≥n cuantitativa precisa a trav√©s del m√©todo TF-IDF.
Por ejemplo, en los primeros registros se observa c√≥mo las frases originales fueron simplificadas a sus componentes l√©xicos m√°s significativos, manteniendo el sentido general pero reduciendo la complejidad sint√°ctica.

Desde una perspectiva metodol√≥gica, este paso es cr√≠tico dentro del pipeline de NLP, ya que determina la calidad de las representaciones vectoriales y, por ende, la capacidad predictiva de los modelos supervisados.
Una normalizaci√≥n consistente reduce la dimensionalidad y mejora la generalizaci√≥n del clasificador, evitando que el modelo aprenda patrones espurios derivados de may√∫sculas, s√≠mbolos o variaciones gramaticales.

En conclusi√≥n, la normalizaci√≥n implementada permiti√≥ obtener un corpus textual sem√°nticamente coherente, compacto y listo para su vectorizaci√≥n, estableciendo las condiciones necesarias para avanzar hacia las siguientes etapas de an√°lisis de sentimiento y modelado con t√©cnicas de Machine Learning y Deep Learning.

```python
# ============================================
# MATRIZ DE CONFUSI√ìN ROBUSTA (AUTO-RECONSTRUCCI√ìN)
# ============================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

def ensure_exists_clean_text_and_sentiment(df):
    # Asegura columnas m√≠nimas
    assert 'clean_text' in df.columns, "Falta 'clean_text'. Ejecut√° la celda de normalizaci√≥n."
    if 'sentiment' not in df.columns:
        # Intentar crearlo desde 'score' si existe
        assert 'score' in df.columns, "Falta 'sentiment' y no existe 'score' para derivarlo."
        max_score = df['score'].max()
        def score_to_sentiment_auto(s, max_score):
            if pd.isna(s): return np.nan
            if max_score > 10:        # escala 0‚Äì100
                if s <= 55: return "negative"
                if s < 70:  return "neutral"
                return "positive"
            else:                     # escala 0‚Äì10
                if s <= 5.5: return "negative"
                if s < 7.0:  return "neutral"
                return "positive"
        df['sentiment'] = df['score'].apply(lambda x: score_to_sentiment_auto(x, max_score))
        print("‚úÖ 'sentiment' creado autom√°ticamente desde 'score'.")
    return df.dropna(subset=['clean_text','sentiment']).copy()

def ensure_model_outputs(globals_dict):
    """
    Si faltan y_test / y_pred / X_test_tfidf, reentrena un baseline r√°pido
    y devuelve (cm, cm_df, y_test, y_pred) listos para graficar.
    """
    g = globals_dict
    if all(k in g for k in ['cm','cm_df','y_test','y_pred']):
        return g['cm'], g['cm_df'], g['y_test'], g['y_pred']

    # Reconstruir pipeline m√≠nimo si faltan variables
    assert 'df' in g, "No encuentro el DataFrame 'df' en el entorno."
    data = ensure_exists_clean_text_and_sentiment(g['df'])

    # Evitar error si alguna clase tiene muy pocos ejemplos
    if data['sentiment'].nunique() < 2:
        raise ValueError("Hay menos de 2 clases en 'sentiment'. Revisa los cortes de score o usa percentiles.")
    class_counts = data['sentiment'].value_counts()
    if (class_counts < 2).any():
        print("‚ö†Ô∏è Alguna clase tiene < 2 ejemplos; ajustando: se quitar√°n clases con 1 ejemplo para evitar errores.")
        keep = class_counts[class_counts >= 2].index
        data = data[data['sentiment'].isin(keep)].copy()

    X_train, X_test, y_train, y_test = train_test_split(
        data['clean_text'], data['sentiment'],
        test_size=0.2, random_state=42, stratify=data['sentiment']
    )

    tfidf = TfidfVectorizer(max_features=20000, ngram_range=(1,2), min_df=2)
    X_train_tfidf = tfidf.fit_transform(X_train)
    X_test_tfidf  = tfidf.transform(X_test)

    logreg = LogisticRegression(max_iter=3000)
    logreg.fit(X_train_tfidf, y_train)
    y_pred = logreg.predict(X_test_tfidf)

    print("\n=== M√âTRICAS (recalculadas) ===")
    print(classification_report(y_test, y_pred, digits=3))

    labels_sorted = sorted(np.unique(y_test))
    cm = confusion_matrix(y_test, y_pred, labels=labels_sorted)
    cm_df = pd.DataFrame(cm, index=labels_sorted, columns=labels_sorted)

    # guardar en globals por si quer√©s usarlos despu√©s
    g.update(dict(tfidf=tfidf, X_test_tfidf=X_test_tfidf, y_test=y_test, y_pred=y_pred, cm=cm, cm_df=cm_df))
    return cm, cm_df, y_test, y_pred

# 1) Asegurar que existen salidas de modelo; si no, se recalculan
cm, cm_df, y_test, y_pred = ensure_model_outputs(globals())

# 2) Graficar heatmap (sin seaborn)
fig, ax = plt.subplots(figsize=(5,4))
im = ax.imshow(cm, interpolation='nearest', cmap='Blues')

ax.set_title("Matriz de confusi√≥n ‚Äì LogReg")
ax.set_xticks(range(len(cm_df.columns)))
ax.set_yticks(range(len(cm_df.index)))
ax.set_xticklabels(cm_df.columns, rotation=45, ha='right')
ax.set_yticklabels(cm_df.index)

for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(j, i, cm[i, j], ha="center", va="center", color="black")

ax.set_xlabel("Predicho")
ax.set_ylabel("Real")
plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
plt.tight_layout()
plt.show()


```

```python
# ==== Reconstrucci√≥n robusta + MLP con fallback ====
import numpy as np, pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report

# 1) Asegurar que existen 'clean_text' y 'sentiment'
assert 'df' in globals(), "No encuentro el DataFrame 'df'."
assert 'clean_text' in df.columns, "Falta 'clean_text'. Ejecut√° la secci√≥n de normalizaci√≥n."
if 'sentiment' not in df.columns:
    assert 'score' in df.columns, "Falta 'sentiment' y no existe 'score' para derivarlo."
    max_score = df['score'].max()
    def score_to_sentiment_auto(s, max_score):
        if pd.isna(s): return np.nan
        if max_score > 10:        # 0‚Äì100
            if s <= 55: return "negative"
            if s < 70:  return "neutral"
            return "positive"
        else:                      # 0‚Äì10
            if s <= 5.5: return "negative"
            if s < 7.0:  return "neutral"
            return "positive"
    df['sentiment'] = df['score'].apply(lambda x: score_to_sentiment_auto(x, max_score))
    print("‚úÖ 'sentiment' creado autom√°ticamente a partir de 'score'.")

data = df.dropna(subset=['clean_text','sentiment']).copy()

# Evitar errores por clases con muy pocos ejemplos
vc = data['sentiment'].value_counts()
if (vc < 2).any():
    keep = vc[vc >= 2].index
    data = data[data['sentiment'].isin(keep)].copy()

# 2) Reconstruir split + TF-IDF si faltan
if 'X_train_tfidf' not in globals() or 'y_train' not in globals() or 'y_test' not in globals():
    X_train, X_test, y_train, y_test = train_test_split(
        data['clean_text'], data['sentiment'],
        test_size=0.2, random_state=42, stratify=data['sentiment']
    )
    tfidf = TfidfVectorizer(max_features=20000, ngram_range=(1,2), min_df=2)
    X_train_tfidf = tfidf.fit_transform(X_train)
    X_test_tfidf  = tfidf.transform(X_test)
    print("üîß Reconstruidos: split y TF-IDF.")

# 3) Intentar MLP con TensorFlow; si no, fallback con sklearn
try:
    import tensorflow as tf
    from tensorflow.keras import layers, models
    from sklearn.preprocessing import LabelEncoder
    from tensorflow.keras.utils import to_categorical
    print("‚úÖ TensorFlow disponible. Entrenando MLP (Keras).")

    le = LabelEncoder()
    y_train_enc = le.fit_transform(y_train)
    y_test_enc  = le.transform(y_test)
    num_classes = len(le.classes_)

    y_train_oh = to_categorical(y_train_enc, num_classes)
    y_test_oh  = to_categorical(y_test_enc, num_classes)

    model = models.Sequential([
        layers.Input(shape=(X_train_tfidf.shape[1],)),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    history = model.fit(
        X_train_tfidf.toarray(), y_train_oh,
        validation_split=0.2, epochs=5, batch_size=64, verbose=1
    )
    test_loss, test_acc = model.evaluate(X_test_tfidf.toarray(), y_test_oh, verbose=0)
    print(f"Test accuracy (MLP-Keras): {test_acc:.3f}")

    import matplotlib.pyplot as plt
    plt.figure(figsize=(6,4))
    plt.plot(history.history['accuracy'], label='train')
    plt.plot(history.history['val_accuracy'], label='val')
    plt.title('Accuracy MLP (Keras)'); plt.xlabel('√âpoca'); plt.ylabel('Accuracy'); plt.legend()
    plt.show()

except Exception as e:
    print("‚ö†Ô∏è TensorFlow no disponible o fall√≥ el entrenamiento. Fallback con sklearn:", e)
    from sklearn.neural_network import MLPClassifier
    mlp = MLPClassifier(hidden_layer_sizes=(256,128), activation='relu', alpha=1e-4,
                        learning_rate_init=1e-3, max_iter=20, random_state=42)
    mlp.fit(X_train_tfidf, y_train)
    y_pred_mlp = mlp.predict(X_test_tfidf)
    acc = accuracy_score(y_test, y_pred_mlp)
    print(f"Test accuracy (MLP sklearn): {acc:.3f}")
    print("\n=== Reporte (MLP sklearn) ===")
    print(classification_report(y_test, y_pred_mlp, digits=3))
```
